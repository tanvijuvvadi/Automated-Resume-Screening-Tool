```python
# ── CELL 5: Text Extraction Module ──────────────────────────

def extract_text_from_txt(file_path):
    """Extract text from plain .txt resume file."""
    try:
        with open(file_path, "r", encoding="utf-8") as f:
            return f.read()
    except Exception as e:
        print(f"❌ Error reading {file_path}: {e}")
        return ""

def extract_text_from_pdf(file_path):
    """Extract text from PDF resume file."""
    import pdfplumber
    text = ""
    try:
        with pdfplumber.open(file_path) as pdf:
            for page in pdf.pages:
                page_text = page.extract_text()
                if page_text:
                    text += page_text + "\n"
    except Exception as e:
        print(f"❌ Error reading PDF {file_path}: {e}")
    return text

def extract_text_from_docx(file_path):
    """Extract text from DOCX resume file."""
    import docx
    text = ""
    try:
        doc = docx.Document(file_path)
        for para in doc.paragraphs:
            text += para.text + "\n"
    except Exception as e:
        print(f"❌ Error reading DOCX {file_path}: {e}")
    return text

def extract_resume_text(file_path):
    """
    Auto-detect file type and extract text.
    Supports: .txt  .pdf  .docx
    """
    ext = os.path.splitext(file_path)[1].lower()
    if ext == ".txt":
        return extract_text_from_txt(file_path)
    elif ext == ".pdf":
        return extract_text_from_pdf(file_path)
    elif ext == ".docx":
        return extract_text_from_docx(file_path)
    else:
        print(f"⚠️ Unsupported format: {ext}")
        return ""

# ── Quick Test ───────────────────────────────────────────────
test_text = extract_resume_text(
    "/content/resume_screener/resumes/resume_alice.txt"
)
print("📄 Sample extracted text (first 300 chars):")
print("-" * 50)
print(test_text[:300])
print("\n✅ Extractor working correctly!")
```

```python
# ── CELL 6: Text Cleaning Module ────────────────────────────
import re

# Common stopwords
STOPWORDS = {
    "a", "an", "the", "and", "or", "but", "in", "on", "at",
    "to", "for", "of", "with", "is", "are", "was", "were",
    "be", "been", "have", "has", "had", "do", "does", "did",
    "will", "would", "can", "could", "should", "may", "might",
    "i", "we", "you", "he", "she", "they", "it", "this",
    "that", "as", "from", "by", "our", "your", "their",
    "not", "also", "using", "used", "including", "experience"
}

def clean_text(text):
    """
    Clean raw text step by step:
    1. Lowercase
    2. Remove special characters
    3. Remove extra whitespace
    4. Remove stopwords
    """
    # Step 1: Lowercase
    text = text.lower()

    # Step 2: Remove special characters (keep only letters + spaces)
    text = re.sub(r'[^a-z\s]', ' ', text)

    # Step 3: Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()

    # Step 4: Remove stopwords and very short words
    words = text.split()
    words = [w for w in words if w not in STOPWORDS and len(w) > 2]

    return " ".join(words)

# ── Quick Test ───────────────────────────────────────────────
raw    = "Alice Johnson has 2+ years of experience in Python, Pandas & ML!"
cleaned = clean_text(raw)

print("Before cleaning:")
print(raw)
print("\nAfter cleaning:")
print(cleaned)
print("\n✅ Cleaner working correctly!")
```

```python
# ── CELL 7: Scoring Module ───────────────────────────────────
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def calculate_similarity_score(resume_clean, jd_clean):
    """
    Calculate how similar a resume is to the job description.
    Uses TF-IDF vectorization + cosine similarity.
    Returns score between 0.0 and 1.0
    """
    corpus = [jd_clean, resume_clean]

    # Build TF-IDF matrix
    vectorizer  = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(corpus)

    # Calculate cosine similarity
    score = cosine_similarity(
        tfidf_matrix[0:1],   # JD vector
        tfidf_matrix[1:2]    # Resume vector
    )[0][0]

    return round(float(score), 4)

def extract_matched_skills(resume_raw, skills_list):
    """
    Check which required skills are present/missing
    in the resume text.
    """
    resume_lower = resume_raw.lower()
    matched = [s for s in skills_list if s.lower() in resume_lower]
    missing = [s for s in skills_list if s.lower() not in resume_lower]
    return matched, missing

# ── Quick Test ───────────────────────────────────────────────
jd_raw    = open("/content/resume_screener/data/job_description.txt").read()
alice_raw = extract_resume_text(
    "/content/resume_screener/resumes/resume_alice.txt"
)

jd_clean    = clean_text(jd_raw)
alice_clean = clean_text(alice_raw)

score = calculate_similarity_score(alice_clean, jd_clean)
print(f"Alice's similarity score: {score}")
print(f"As percentage: {score * 100:.1f}%")
print("\n✅ Scorer working correctly!")
```

```python
# ── CELL 8: Ranking and Shortlisting Module ──────────────────

def rank_candidates(results, threshold=0.40):
    """
    Sort candidates by score (highest first).
    Apply shortlist/reject decision based on threshold.
    """
    # Sort by score descending
    ranked = sorted(results, key=lambda x: x['score'], reverse=True)

    # Assign rank and status
    for i, candidate in enumerate(ranked):
        candidate['rank'] = i + 1
        if candidate['score'] >= threshold:
            candidate['status'] = "Shortlisted ✅"
        else:
            candidate['status'] = "Rejected ❌"

    return ranked

print("✅ Ranker ready!")
```

```python
# ── CELL 9: Report Generation Module ────────────────────────
import pandas as pd

def generate_report(ranked_results,
                    output_path="/content/resume_screener/outputs/screening_report.csv"):
    """
    Convert ranked results to DataFrame and save as CSV.
    """
    os.makedirs("/content/resume_screener/outputs", exist_ok=True)

    df = pd.DataFrame(ranked_results)

    # Select and order columns
    columns = ['rank', 'candidate', 'score', 'score_percent',
               'status', 'matched_skills', 'missing_skills']
    df = df[[c for c in columns if c in df.columns]]

    df.to_csv(output_path, index=False)
    print(f"✅ Report saved to: {output_path}")
    return df

print("✅ Reporter ready!")
```

```python
# ── CELL 10: MAIN RUNNER — Run the full pipeline ─────────────

RESUMES_FOLDER = "/content/resume_screener/resumes/"
JD_FILE        = "/content/resume_screener/data/job_description.txt"
SKILLS_FILE    = "/content/resume_screener/data/skills_list.txt"
THRESHOLD      = 0.40

print("=" * 55)
print("   🤖 AUTOMATED RESUME SCREENING TOOL")
print("=" * 55)

# Load JD and skills
jd_raw      = open(JD_FILE, encoding="utf-8").read()
jd_clean    = clean_text(jd_raw)
skills_list = [s.strip() for s in
               open(SKILLS_FILE).read().splitlines() if s.strip()]

print(f"\n📋 Job Description loaded.")
print(f"🔑 Required Skills ({len(skills_list)}): {skills_list}")

# Get all resume files
resume_files = [
    f for f in os.listdir(RESUMES_FOLDER)
    if f.endswith((".txt", ".pdf", ".docx"))
]

print(f"\n📂 Found {len(resume_files)} resumes to screen\n")

results = []

for i, filename in enumerate(resume_files):
    filepath = os.path.join(RESUMES_FOLDER, filename)
    print(f"🔍 [{i+1}/{len(resume_files)}] Processing: {filename}")

    # Extract and clean
    raw_text    = extract_resume_text(filepath)
    clean       = clean_text(raw_text)

    # Score
    score       = calculate_similarity_score(clean, jd_clean)

    # Skill matching
    matched, missing = extract_matched_skills(raw_text, skills_list)

    results.append({
        "rank"          : 0,
        "candidate"     : filename,
        "score"         : score,
        "score_percent" : f"{score * 100:.1f}%",
        "matched_skills": ", ".join(matched),
        "missing_skills": ", ".join(missing),
        "status"        : ""
    })

# Rank and shortlist
ranked = rank_candidates(results, threshold=THRESHOLD)

# Print results
print("\n" + "=" * 55)
print("📊 SCREENING RESULTS")
print("=" * 55)

for r in ranked:
    print(f"\n{'='*40}")
    print(f"  #{r['rank']}  {r['candidate']}")
    print(f"  Score         : {r['score_percent']}")
    print(f"  Status        : {r['status']}")
    print(f"  Matched Skills: {r['matched_skills'] or 'None'}")
    print(f"  Missing Skills: {r['missing_skills'] or 'None'}")

# Generate report
print("\n" + "=" * 55)
df = generate_report(ranked)
print("\n📁 Full Report Preview:")
print(df.to_string(index=False))
```
